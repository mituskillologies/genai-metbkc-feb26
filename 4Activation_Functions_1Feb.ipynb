```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.25)
```

```python
x
```

```python
def sigmoid(x):
  return 1/ (1 + np.exp(-x))
```

```python
y = sigmoid(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
def relu(x):
  return np.maximum(0, x)
```

```python
y = relu(x)
```

```python
plt.plot(x, y)
```
